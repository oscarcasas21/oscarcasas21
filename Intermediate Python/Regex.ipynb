```python
import pandas as pd
import re
import numpy as np
```

```python
JP_addresses = pd.read_csv('/Users/oscar/Desktop/Exercise4_JP_addresses.csv')
JP_addresses.head()
```

```python
zipcodes=[]
for i in  JP_addresses.index:
    x = re.findall('\d{3}[-.\s]\d{4}', str(JP_addresses['address'][i]))
    if len(x) > 0:
        zipcodes.append(x[0])
    else:
        zipcodes.append('empty')

JP_addresses['zipcode'] = zipcodes
JP_addresses['zipcode']
```

```python
street=[]
for i in  JP_addresses.index:
    split_data = str(JP_addresses['address'][i]).split(",")
    if split_data[0]=='Japan':
        street.append(split_data[-1])
    else:
        street.append(split_data[0])

JP_addresses['Street'] = street
JP_addresses['Street']

JP_addresses
```
